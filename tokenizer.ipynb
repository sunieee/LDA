```python
import jieba
import jieba.posseg as pseg
from collections import Counter
import os
import re

dirname = '张学良'
# dirname = '文章'

# typ = 'sentence'
typ = 'row'
# typ = 'passage'


jieba.load_userdict("用户词典.txt") 
input_path=f'{dirname}/全文.txt'
output_path=f'{dirname}/全文分词.txt'
```

```python
# 设置停用词
print('start read stopwords data.')
stopwords = set()

for file in os.listdir('stopwords'):
    with open(f'stopwords/{file}', 'r', encoding='utf-8') as f:
        for line in f:
            if len(line)>0:
                stopwords.add(line.strip())
            
stopwords = list(stopwords)
print(len(stopwords), stopwords)
```

```python
all_cuts = []
all_words = []

def tokenizer(s):
    words = []
    cut = list(pseg.cut(s))
    for word in cut:
        word_txt = word.word
        if word_txt not in stopwords and len(word_txt) > 1 and \
                not re.match(r'[0-9]+', word_txt) and \
                word_txt not in ['\n', '\u3000', ' ', '\t']:
#            if word.flag=='nr':
            words.append(word_txt)
#            words.append(word.flag)
    print(len(cut), len(words))
    all_cuts.append(len(cut))
    all_words.append(len(words))
    return words
```

```python
lines = []

if typ == 'sentence':
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            for sentence in line.split('。'):
                if len(sentence):
                    s=tokenizer(sentence.strip())
                    lines.append(" ".join(s))
                    cnt.append(len(s))

if typ == 'row':
    with open(input_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if len(line):
                s=tokenizer(line)
                lines.append(" ".join(s))
                cnt.append(len(s))

def isint(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

if typ == 'passage':
    for file in os.listdir(dirname):
        if not os.path.isdir(file) and os.path.splitext(file)[1] == '.txt' and isint(file[0]):
            with open(f'{dirname}/{file}', 'r', encoding='utf-8') as f:
                line = f.read()
                line = line.strip()
                if len(line):
                    s=tokenizer(line)
                    lines.append(" ".join(s))
                    cnt.append(len(s))
```

```python
import numpy as np
print('行数', len(all_cuts))
print('每行词数', np.mean(all_cuts), '±', np.std(all_cuts))
print('每行有效词数', np.mean(all_words), '±', np.std(all_words))

with open(output_path,'w',encoding='utf-8') as f:
    f.write('\n'.join(lines))
```
